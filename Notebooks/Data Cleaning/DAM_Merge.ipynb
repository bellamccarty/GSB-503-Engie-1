```python
import pandas as pd
import glob
import os
```

```python
folder_path = r'C:\Users\isabe\PycharmProjects\GSB-503-Engie-1\Data\DAM 60d ERCOT'
csv_files = glob.glob(folder_path + '/*.csv')

# Dictionary to store DataFrames with unique names
dataframes = {}

# Iterate over the list of CSV files
for i, file in enumerate(csv_files, start=1):
    # Read the CSV file
    df = pd.read_csv(file)
    
    # Extract the base name of the file (without path and extension)
    base_name = os.path.basename(file).replace('.csv', '')
    
    # Create a unique name for the DataFrame
    unique_name = f"df_{base_name}"
    
    # Store the DataFrame in the dictionary
    dataframes[unique_name] = df

for name, df in dataframes.items():
    print(f"{name} shape: {df.shape}")
```

```python
offer_awards = dataframes['df_dam_energyonlyofferawards_60d']
energy_offers = dataframes['df_dam_energyonlyoffers_60d']
energy_offers.rename(columns={'energy_only_offer_id': 'offer_id'}, inplace=True)

merged_data = offer_awards.merge(
    energy_offers,
    on=['delivery_date', 'hour_ending', 'settlement_point', 'qse_name', 'p_posted_date', 'offer_id'],
    how='outer'
)
```

```python
energy_bids = dataframes['df_dam_energybids_60d']

merged_data = merged_data.merge(
    energy_bids,
    on=['delivery_date', 'hour_ending', 'settlement_point', 'qse_name', 'p_posted_date',
        'multi_hour_block_indicator', 'block/curve_indicator'],
    how='outer'
)

merged_data.head()
```

```python
gen_resource_data = dataframes['df_dam_gen_resource_data_60d']
gen_resource_data.rename(columns={'settlement_point_name': 'settlement_point', 'qse': 'qse_name'}, inplace=True)

merged_data = merged_data.merge(
    gen_resource_data,
    on=['delivery_date', 'hour_ending', 'qse_name', 'settlement_point',
        'p_posted_date'],
    how='outer'
)
```

```python
gen_resource_data_as = dataframes['df_dam_generation_resource_asoffers_60d']
gen_resource_data_as.rename(columns={'generation_resource_name': 'resource_name', 'qse': 'qse_name'}, inplace=True)

merged_data = merged_data.merge(
    gen_resource_data_as,
    on=['delivery_date', 'hour_ending', 'qse_name', 'resource_name', 'dme',
        'p_posted_date'],
    how='left'
)
```
